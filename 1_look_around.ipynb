```python
def plot_sunburst(df, root_field, branch_field="Transported"):
    fig = px.sunburst(
        data_frame=df,
        title=root_field,
        path=[root_field, branch_field],  # Root, branches, leaves
        color=branch_field,
        # color_discrete_sequence=px.colors.qualitative.Pastel,
    )
    fig.update_traces(textinfo='label+percent entry')
    fig.update_layout(width=600, height=600)
    fig.show()
```

```python
def pie_field_count(df, field, title):
    g_df = pd.DataFrame(df.groupby(field).size()).reset_index()
    g_df.rename(columns={0:'count'}, inplace=True)

    chart = px.pie(g_df, 
        values="count", 
        names=field, 
        title=title,
        width=800,                          #figure width in pixels
        height=600,                         #figure height in pixels
        hole=0.5,                           #represents the hole in middle of pie
        )
    chart.update_traces(
        textposition='outside', 
        textinfo='percent+label',
        marker=dict(line=dict(color='#000000', width=3)),
        # pull=[0, 0, 0.2, 0], 
        opacity=0.7, 
        # rotation=180,
        )

    chart.show()
```

```python
pie_field_count(df, "group_num", "Members per GROUP_NUM")
```
